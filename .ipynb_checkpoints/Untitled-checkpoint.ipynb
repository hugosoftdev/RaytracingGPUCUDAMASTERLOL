```python
import numpy as np
import matplotlib.pyplot as plt

def plot_double_axis(x , data1, data2):
    t = x
    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('exp', color=color)
    ax1.plot(t, data1, color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
    ax2.plot(t, data2, color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
```

```python
plot_double_axis(resolucao, render_time, create_word)
```

```python
plt.title("Tempo de execução em GPU das funções do programa")
plt.plot(resolucao, render_time,  '-b', label='render')
plt.plot(resolucao, create_word,  '-r', label='create_word')
plt.xlabel("Resolução em escala 10 a sétima")
plt.ylabel("Tempo em segundos")
plt.legend(loc='best')
plt.show()
```
