```python
import numpy as np
resolucao = [400*200, 1200*600, 3600*1800, 10800*5400]
render_time = [213* (10**-6), 1.83*(10**-3), 16*(10**-3), 129*(10**-3)]
render_time_percentage = [77.75, 96.8, 99.63,99.95]
create_word = [61*(10**-6),60*(10**-6), 61.64*(10**-6), 61.54*(10**-6)]
create_word_percentage = [22.25, 3.20, 0.37,0.05]
cuda_malloc_managed = [140*(10**-3),143*(10**-3), 172*(10**-3), 405*(10**-3) ]
cuda_malloc_managed_percentage = [70,69,60,46.6]
cuda_device_reset = [51*(10**-3),53.4*(10**-3), 85.3*(10**-3), 328*(10**-3)] 
cuda_device_reset_percentage = [25,26,30,37]
cuda_device_sync_percentage = [0.15, 1, 6, 16]
gpu_total = [0.18,0.27,1.7,12.967]
cpu_vetorizado = [0.02,0.17,1.55,12.41]
cpu_cru = [0.05,0.43,3.84,30.05]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.title("Tempo de execução em GPU")
plt.plot(resolucao, render_time,  '-b', label='render')
plt.plot(resolucao, create_word,  '-r', label='create_word')
plt.xlabel("Resolução em escala 10 a sétima")
plt.ylabel("Tempo em segundos")
plt.legend(loc='best')
plt.show()
```

```python
plt.title("Tempo de execução em GPU de chamadas de API do cuda")
plt.plot(resolucao, cuda_malloc_managed,  '-b', label='cuda_malloc_managed')
plt.plot(resolucao, cuda_device_reset,  '-r', label='cuda_device_reset')
plt.xlabel("Resolução em escala 10 a sétima")
plt.ylabel("Tempo em segundos")
plt.legend(loc='best')
plt.show()
```

```python
plt.title("Tempo percentual de  execução em GPU de chamadas de API do cuda")
plt.plot(resolucao, cuda_malloc_managed_percentage,  '-b', label='cuda_malloc_managed')
plt.plot(resolucao, cuda_device_reset_percentage,  '-r', label='cuda_device_reset')
plt.plot(resolucao, cuda_device_sync_percentage,  '-g', label='cuda_device_sync')
plt.xlabel("Resolução em escala 10 a sétima")
plt.ylabel("Tempo percentual")
plt.legend(loc='best')
plt.show()
```

```python
plt.title("Tempo de execução total do programa")
plt.plot(resolucao, gpu_total,  '-b', label='GPU')
plt.plot(resolucao, cpu_vetorizado,  '-r', label='Com Vetorização')
plt.plot(resolucao, cpu_cru,  '-g', label='Sem Vetorização')
plt.xlabel("Resolução em escala 10 a sétima")
plt.ylabel("Tempo em segundos")
plt.legend(loc='best')
plt.show()
```
